2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
ax = sns.pointplot(x='data_ref',y='posse_de_veiculo',hue='sexo', data=base, dodge=True, ci=95)

plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.figure(figsize=(8, 6))
plt.show()
````

* Observa-se uma tendência estável de homens possuírem veículos ao longo do tempo, e mulheres, não. Por sua estabilidade, seria possível usar essas variáveis em uma modelagem.

````python
ax = sns.pointplot(x='data_ref',y='posse_de_imovel',hue='sexo', data=base, dodge=True, ci=95)

plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.figure(figsize=(8, 6))
plt.show()
````

* Apesar de menor estabilidade, homens consistentemente apresentam a posse de imóvel de forma mais frequente que e mulheres. Possivelmente seria necessário revisar o modelo após um período de tempo devido ao desvio apresentado no período mais recente.

````python
ax = sns.pointplot(x='data_ref',y='idade',hue='educacao', data=base, dodge=True, ci=95)

plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.figure(figsize=(8, 6))
plt.show()
````

* Parece que há uma tendência de pessoas mais velhas terem menos educação, evidente pelo grupo entre 45-55 anos de idade terem a maior concentração de pessoas que só completaram o primário e o grupo de 45 anos de idade ter completado apenas o secundário, tendo em contraste o grupo de 30-35 anos demonstrarem uma pequena consistência em possuírem pós graduação. As variáveis parecem consistentes o bastante, mas os picos e vales em Pós Graduação poderiam ser um problema - Talvez seja necessário remover esse fator para uma boa modelagem.

````python
ax = sns.pointplot(x='data_ref',y='idade',hue='estado_civil', data=base, dodge=True, ci=95)

plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.figure(figsize=(8, 6))
plt.show()
````

* Essas variáveis parecem bem consistentes, e boas para um modelo, apresentando mínima variação ao longo do tempo.

````python
ax = sns.pointplot(x='data_ref',y='posse_de_imovel',hue='tipo_renda', data=base, dodge=True, ci=95)

plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.figure(figsize=(8, 6))
plt.show()
````

* Pode-se observar que Pensionistas tem a tendência maior de possuírem imóvel, e há uma distribuição bem equilibrada entre todas as outras categorias logo abaixo. Há boa estabilidade, apesar de que o fator dos Bolsistas poder interfirir na modelagem.
